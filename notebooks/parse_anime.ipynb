## 1. Data preparation
### Getting List of All Anime on File

```python
import requests
import time

BASE_URL = "https://api.jikan.moe/v4/anime"

all_anime = []
page = 1

while True:
    if page % 3 == 0:
        time.sleep(3)
        
    print(f"...Adding Page {page}...")
        
    resp = requests.get(BASE_URL, params={"page": page}) #"limit": 25})
    resp.raise_for_status()
    payload = resp.json()

    anime_list = payload["data"]
    
    all_anime.extend(anime_list) # essentially all_anime += anime_list
    
    print(f"[...Page {page} added!...]")
    
    if not payload["pagination"]["has_next_page"]:
        break

    page += 1

print(f"Fetched {len(all_anime)} anime")
```

### Saving results to a json file

```python
import json

# Define the filename
filename = "all_anime.json"

# Open the file in write mode ('w') and use json.dump()
with open(filename, 'w') as f:
    json.dump(all_anime, f, indent=4) # Using indent for human-readable formatting

print(f"Array successfully saved to {filename}")

```

### Extracting Core Anime Fields
